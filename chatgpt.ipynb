```python
import numpy as np
import matplotlib.pyplot as plt

# グリッドの作成
w1, w2 = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))

# L1正則化の等高線
l1 = np.abs(w1) + np.abs(w2)

# L2正則化の等高線
l2 = w1**2 + w2**2

# プロットの設定
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# L1正則化のプロット
ax1.contour(w1, w2, l1, levels=np.logspace(-1, 1, 10))
ax1.set_title('L1 Regularization (Lasso)')
ax1.set_xlabel('w1')
ax1.set_ylabel('w2')

# L2正則化のプロット
ax2.contour(w1, w2, l2, levels=np.logspace(-1, 1, 10))
ax2.set_title('L2 Regularization (Ridge)')
ax2.set_xlabel('w1')
ax2.set_ylabel('w2')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso, Ridge

# データセットの作成
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# グラフ表示のためにデータポイントをプロット
plt.scatter(X, y, alpha=0.5)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Generated Data")
plt.show()

# 通常の線形回帰
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)

# L1正則化（Lasso）
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)
y_pred_lasso = lasso_reg.predict(X)

# L2正則化（Ridge）
ridge_reg = Ridge(alpha=1)
ridge_reg.fit(X, y)
y_pred_ridge = ridge_reg.predict(X)

# 結果のプロット
plt.scatter(X, y, alpha=0.5, label="Data")
plt.plot(X, y_pred_lin, color='blue', label="Linear Regression")
plt.plot(X, y_pred_lasso, color='red', label="Lasso Regression (L1)")
plt.plot(X, y_pred_ridge, color='green', label="Ridge Regression (L2)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear vs Lasso vs Ridge Regression")
plt.show()

# パラメータの表示
print("Linear Regression coefficients:", lin_reg.coef_)
print("Lasso Regression coefficients:", lasso_reg.coef_)
print("Ridge Regression coefficients:", ridge_reg.coef_)
```
